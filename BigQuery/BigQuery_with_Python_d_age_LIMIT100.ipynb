```python
project_id = 'crucial-bonsai-377206'
```

```python
from google.cloud import bigquery
```

```python
client = bigquery.Client(project=project_id)
```

```python
client
```

```python
row_count = client.query('''
  SELECT
    COUNT(*) as total
  FROM `crucial-bonsai-377206.imbd_dage.d_age`
  LIMIT 100''')
```

```python
row_count
```

```python
row_count.result()
```

```python
for r in row_count.result():
    print(r)
```

```python
df = row_count.to_dataframe()
```

```python
df
```

```python
d_age = client.query('''
SELECT
    primary_name, d_age, primary_profession, title
FROM
    `crucial-bonsai-377206.imbd_dage.d_age`
LIMIT 100''')
```

```python
d_age
```

```python
d_age.result()
```

```python
for r in d_age.result():
  print(r)
```
